### 2. 인기 있는 영화 찾기 

- 기준 : 평점이 4.5 이상이고 평점 개수가 500개 이상인 영화

```python
# 평점이 높은 순서로 10개 정렬(head 활용)
data2.sort_values('평점', ascending=False).head(10)
```

```python
# 하나의 영화에 여러개의 평점이 존재하는 1:N 관계이므로 
# 사용자 아이디는 사실상 영화별 평균평점을 구하는데는 불필요한 자료임
# 따라서 내가 원하는 로우(index) : 영화 제목
# 원하는 컬럼(values) : 영화 평점
# 집계함수(aggfunc) : 평균 
# 으로 pivot_table을 data2에 걸어 집계해보세요. 정렬은 하지 않아도 좋습니다.
data3 = data2.pivot_table(index='제목', aggfunc='mean', values='평점')
data3
```

```python
# data3를 평점 높은순으로 정렬해주세요.
data3.sort_values(by='평점', ascending=False).head(10)
```

```python
# 평점을 기준으로 정렬하되 평균평점만 출력하는게 아니라, 평균갯수도 같이 출력
# data2에 대해서 pivot_table을 걸 때
# 로우 : 제목
# 값 : 평점
# 집계함수 : ['mean', 'count']
result = data2.pivot_table(index='제목', aggfunc=['mean', 'count'], values='평점')
result
```

```python
# result에 대해 swaplevel()을 이용해 계층색인을 재배치해주세요.
result2 = result.swaplevel(1, 0, axis=1)
```

```python
result2.head()
```

```python
# 위 자료를 평점, mean 순으로 정렬해주시되, 평점이 동점이면 평점, count가 많은게 위로 가게 해주세요.
# [('평점', 'mean'), ]
result3 = result2.sort_values(by=[('평점', 'mean'), ('평점', 'count')], ascending=False)
result3
```

```python
# result3에 대해서 조건색인을 이용해 
# 평균평점이 4.5이상이면서 평점 개수가 500개 이상인 영화만 남겨보기
# ex) 이중으로 부여된 컬럼명은 result3['평점']['mean']과 같이 조회한다.
# 변수 final_result에 저장해주세요.
# 조건색인 = 변수명[변수명[컬럼]에대한조건식]
final_result = result3[(result3['평점']['mean'] >= 4.5) & (result3['평점']['count'] >= 500)]
final_result
```
