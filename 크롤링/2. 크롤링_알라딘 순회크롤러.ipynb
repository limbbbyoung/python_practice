```python
# 크롤링 작업을 위한 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# 셀레니움 기본세팅을 해주세요 .
# 알라딘 베스트셀러 페이지에 접속하도록 만들어주시면 됩니다.

# browser 라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
browser= webdriver.Chrome('chromedriver.exe') # 확장자를 빼도 잘 돌아감

# .get(접속주소) 를 입력하면 해당 주소로 브라우저 창이 이동합니다.
browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&page=1")

# 소스코드 변수에 저장하기
# 브라우저가 특정 페이지에 접근했을때, 해당 페이지 소스코드 전체 긁어오기
# 여기까지가 셀레니움의 역할임. 가져온 소스코드는 뷰티풀수프로 정제함
# browser.page_source는 source 변수에 전체 페이지 소스를 문자로 저장함
source = browser.page_source
print(type(source))
print(source)
```

```python
# 소스코드 파싱
aladin_code = BeautifulSoup(source, "html.parser")
print(type(aladin_code))
```

```python
# 순위별로 데이터를 가져오기 위해 처리
# 태그명이 div 이면서 클래스명이 ss_book_box인 요소를 가져와야함
div_box_list = aladin_code.find_all("div", class_="ss_book_box")
print(len(div_box_list))
```

```python
for div in div_box_list:
    li_list = div.find_all("li")
    print(len(li_list))
    # 해당 li_list 중에 우리가 찾는 제목도 있고, 가격도 있고, 저자도 있음
    # li의 길이, len(li_list)가 7인 경우 제목이 짤려보이는 상황이기 때문에
    # 7인 경우만 0번째 인덱스부터 보여줄 수 있도록 설정
    if (len(li_list)==7) :
        print(li_list[0].text)
        print(li_list[1].text)
        print(li_list[2].text)
    else :
        print(li_list[1].text)
        print(li_list[2].text)
        print(li_list[3].text)
    print("----------------------------")
```

```python
# browser 변수를 이용해서 알라딘 베스트셀러 페이지를 열어주시고
# 그 와중에 각 페이지 이동당 2초씩 딜레이를 주시면 됩니다.
# 반복문을 이용해 마지막 페이지까지 순회시켜주세요.

# browser 라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
browser= webdriver.Chrome('chromedriver.exe') # 확장자를 빼도 잘 돌아감

# 책 순위를 찍습니다.
book_rank = 0

# 시작 페이지 
page = 1
while page < 21:
    browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1"%page)
    page += 1
    time.sleep(3)
    
    # 소스코드 저장
    source = browser.page_source
    
    # 파싱 
    aladin_code = BeautifulSoup(source, "html.parser")
    
    # 출력
    div_box_list = aladin_code.find_all("div", class_="ss_book_box")

    #li_list의 몇 번째가 제목이고, 가격이고, 저자인지 알면 교보처럼 출력하는데 어려움이 없음
    for div in div_box_list:
        # 책이 바뀔때마다 책 순위를 1위씩 더해줌
        book_rank += 1 # 북 랭크를 표시해줌으로써 책의 순위를 나타낼 수 있음
        li_list = div.find_all("li")
        # 해당 li_list 중에 우리가 찾는 제목도 있고, 가격도 있고, 저자도 있음
        # li의 길이, len(li_list)가 7인 경우 제목이 짤려보이는 상황이기 때문에
        # 7인 경우만 0번째 인덱스부터 보여줄 수 있도록 설정
        if (len(li_list) == 7) :
            print(li_list[0].text)
            title_list.append(li_list[0].text)
            print(li_list[1].text)
            author_list.append(li_list[1].text)
            print(li_list[2].text)
            price_list.append(li_list[2].text)
        else :
            print(li_list[1].text)
            title_list.append(li_list[1].text)
            print(li_list[2].text)
            author_list.append(li_list[2].text)
            print(li_list[3].text)
            price_list.append(li_list[3].text)
        print("----------------------------")
```

```python
# 제목, 가격, 저자 정보를 저장할 리스트 생성
title_list = []
price_list = []
author_list = []
```

```python
# 위의 txt양식을 바꿔서 
# 제목, 저자, 가격
# 형식으로 컬럼을 세팅해 csv로 뽑아보겠습니다.
# 가격에 "," 가 포함되어있으므로, 깨지지 않게 처리하기위해 ,를 없애고 txt파일로 써주셔야합니다.

f = codecs.open("c:/crawler2/best1000.txt", mode='w', encoding='utf-8')

# 데이터 추가
for i in range(0, len(title_list)):
    # 각 책에 대한 정보 write로 한 줄에 적기
    f.write(title_list[i].replace(",","") + "," +
            author_list[i].replace(",","") + "," +
            price_list[i].replace(",","") + "\n")   

# 파일 연결을 마지막에 끊어줘야 삭제나 수정, 이동가능
f.close()
```
