### 크롤링

- 웹에서 원하는 자료를 컴퓨터에게 수집해오도록 명령내리는 기술
- requests 라이브러리를 활용한 다이렉트 크롤링
- selenium 라이브러리를 활용한 간접 크롤링 등이 있다.
- 크롤러의 역할은 원하는 정보를 가져오는 것 까지이며
- 실제 원하는 정보를 용도에 맞게 가공하는것은 beautifulsoup가 담당한다.

```python
# 크롤링 작업을 위한 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
```

# selenium 설치

1. Anaconda Navigator의 좌측 Environments 클릭
2. 가운데 base(root) 사이에 있는 ▶ 클릭 후 open terminal 클릭

```python
# browser 라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
browser= webdriver.Chrome('chromedriver.exe')


# .get(접속주소) 를 입력하면 해당 주소로 브라우저 창이 이동합니다.
browser.get("https://www.naver.com")


# 코드 실행간에 텀 주기(3초)
time.sleep(3)


# '교보문고' 키워드 입력
search_query = browser.find_element(By.XPATH, '//*[@id="query"]')
search_query.send_keys('교보문고')

time.sleep(1)

# 검색 버튼 클릭
browser.find_element(By.XPATH, '//*[@id="search_btn"]').click()

time.sleep(2)

# '교보문고' 링크 클릭
browser.find_element(By.XPATH, '//*[@id="main_pack"]/section[1]/div/div/div[1]/div/div[1]/a/span[2]').click()

time.sleep(10)

browser.close()

# 전체적으로 참고가 가장 잘 될만한 문서는 공식문서를 참고하면 됨.
# 공식문서에서 왠만한 답은 다 얻을 수 있음
```

### 특정 url로 접근했을때 바로 원하는 정보를 얻을 수 있는 경우

- 그냥 바로 접근하면 됨(굳이 복잡하게 검색하고 나온 결과를 어쩌고..할 필요는 없음)
- 그러나 페이스북 같이 특정 조건을 만족해야 추가적인 자료를 보여주는 사이트도 있고
- 로그인해야만 자료에 접근할 수 있는 사이트도 있기 때문에
- 어떻게 접근해야 원하는 자료를 얻어올 수 있는지는 신중하게 고려해야함

```python
# 타켓 주소 : http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?orderClick=d79
# 브라우저가 해당 주소에 접속하도록 처리
browser= webdriver.Chrome('chromedriver.exe')

browser.get("http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?orderClick=d79")

time.sleep(2)

# 브라우저가 특정 페이지에 접근했을때, 해당 페이지 소스코드 전체 긁어오기
# 여기까지가 셀레니움의 역할임. 가져온 소스코드는 뷰티풀수프로 정제함
# browser.page_source는 source 변수에 전체 페이지 소스를 문자로 저장함
source = browser.page_source
print(type(source))
print(source)
```

```python
# BeautifulSoup(소스코드, "html.parser") 로 입력시 해당 코드를 html 형식으로 인식함
# 해당 str(문자열) 자료를 html형식으로 인식시키는 것을 "파싱" 이라고 합니다.
# 책을 소유하고 있다고 그 책에 대해 반드시 이해하고 있는건 아니듯 책 내용을 읽어서 이해하고
# 지식을 구조화해야 비로소 책을 이해하고 있다고 할 수 있는데
# 뷰티플수프도 소스코드가 단순 문자로 존재할때는 기능을 쓸 수 없지만
# 파싱을 통해 소스코드 구조를 이해하면 정제를 매우 빠르고 쉽게 처리해줍니다.
parsed_source = BeautifulSoup(source, "html.parser")
print(type(parsed_source))
print(type(source))
```

```python
# 제목 가져오기 
# 파싱된코드.find_all("태그명", class="클래스명") 입력시 해당 태그나 클래스에 대한 데이터만 가져올 수 있음
div_tag_title_list = parsed_source.find_all("div", class_="title")
print(len(div_tag_title_list)) # 현재 "div" 태그이면서 class_가 title인 요소는 전체 페이지에서 82개가 있음
```

```python
# 제목이 82개이므로 다시 범위를 좁혀서 가져오기
div_tag_title_list = parsed_source.find_all("ul", class_="list_type01")
len(div_tag_title_list)
ul_tag_type01 = div_tag_title_list[0] # 리스트 속에 담긴 1개의 태그이므로 1개라고 해도 바로 명령을 받지 못함.
                    # 인덱싱을 통해 무조건 꺼내놓고 명령을 내려야 합니다.
div_tag_title_list = ul_tag_type01.find_all("div", class_="title")
len(div_tag_title_list)
# 웹 페이지에서 가지고 오자고 하는 데이터를 범위를 좁혀서 가지고 오는 것이 크롤러에서 제일 힘든 점임
```

```python
# 50개 데이터를 반복문으로 돌려서 제목만 남기기
# .text는 태그 데이터를 다 제거해서 노출된 문자만 남깁니다.
# 이제 개행문자나 공백 등을 없애서 글자만 정확히 남겨보세요
for item in div_tag_title_list:
    print(item.text.strip())
    title_list.append(item.text.strip())
    print("-----------------------")
```

```python
# 이번에는 저자만 50개를 출력해주세요.
div_tag_title_list = parsed_source.find_all("ul", class_="list_type01")
len(div_tag_title_list)
ul_tag_type01 = div_tag_title_list[0] # 리스트 속에 담긴 1개의 태그이므로 1개라고 해도 바로 명령을 받지 못함.
                    # 인덱싱을 통해 무조건 꺼내놓고 명령을 내려야 합니다.
div_tag_author_list = ul_tag_type01.find_all("div", class_="author")
len(div_tag_author_list)

for item in div_tag_author_list:
    text_list = item.text.split("|")
    print(item.text.strip().replace('\n',"").replace("\t","").split("|")[0].replace("저자 더보기", ""))
    author_list.append(item.text.strip().replace('\n',"").replace("\t","").split("|")[0].replace("저자 더보기", ""))
    print("-----------------------")
```

```python
# 이번에는 가격만 50개를 출력해주세요.
div_tag_title_list = parsed_source.find_all("ul", class_="list_type01")
len(div_tag_title_list)
ul_tag_type01 = div_tag_title_list[0] # 리스트 속에 담긴 1개의 태그이므로 1개라고 해도 바로 명령을 받지 못함.
                    # 인덱싱을 통해 무조건 꺼내놓고 명령을 내려야 합니다.
div_tag_price_list = ul_tag_type01.find_all("div", class_="price")
len(div_tag_price_list)

for item in div_tag_price_list:
    print(item.text.strip().replace('\n',"").replace("\t","").replace(",",""))
    price_list.append(item.text.strip().replace('\n',"").replace("\t","").replace(",",""))
    print("-----------------------")
```

### codecs 라이브러리 

- 파이썬 3.5버전 들어서 내장 라이브러리로 바뀜(예전에는 cmd창에서 pip install codecs로 설치해야 했음)
- 파이썬으로 텍스트파일을 제어할 수 있도록(읽어오기, 쓰기 등) 도와줌
- 콘솔창에 출력된 내용을 txt파일로 옮겨서 출력할 때 사용
<br><br>
- 특이사항 
    - 개행은 \r\n으로 처리함
    - mode => w(기존에 있던 자료를 없애고 새로 파일 입력)
    - mode => a(기존에 있던 자료에 이어서 계속 입력)
    - mode => r(텍스트파일에 있던 내용 읽어오기)

```python
# 라이브러리 임포트
import codecs
```

```python
# f변수가 텍스트변수 그 자체처럼 사용중
# .open(파일경로(파일이 해당 경로에 없으면 생성), encoding="인코딩방식", mode="모드")
# w = write(글쓰기), a=append(기존 내역에 추가), r:read, 기본 파일 읽어서 파이썬 내부 문자로 바꾸기
f = codecs.open("c:/crawler2/test.txt", mode='w')

# Hello world를 10번 적어주세요
for i in range(1, 11): # 적는 숫자가 곧 돌아가는 횟수 ex) range(10)
    f.write("Hello world \n")

# 파일 연결을 마지막에 끊어줘야 삭제나 수정, 이동가능
f.close()
```

```python
# 기본 텍스트자료 읽어오기(f변수에 r모드로 읽어서 넣어보세요.)
f = codecs.open("c:/crawler2/test.txt", mode = 'r', encoding='utf-8')

# f.read()를 이용해 전체 문자를 읽어올 수 있음
# print(f.read())

# .readlines()로 가져오면 알아서 한줄 단위로 끊어서 가져옵니다.
print(f.readlines())

f.close()
```

```python
# 상단의 도서 목록을 txt형식으로 저장해주세요.
# c:/crawler2/result.txt 로 저장해주시면 됩니다.

f = codecs.open("c:/crawler2/result.txt", mode='w')

# 데이터 추가
for i in range(0, len(title_list)):
    f.write(title_list[i] + "\n" +
            author_list[i] + "\n" +
            price_list[i] + "\n-----------------------\n")   

# 파일 연결을 마지막에 끊어줘야 삭제나 수정, 이동가능
f.close()
```

# CSV(excel)형식으로 뽑기

- csv는 ,로 텍스트파일을 구분해 만든 다음
- .csv로 확장자를 바꾸면 엑셀형식으로 나열된 데이터가 나옵니다.
- 그래서 , 를 이용해 구분한다는 점 때문에 , 가 들어간 데이터에는 쓸 수 없습니다.

```python
# 위의 txt양식을 바꿔서 
# 제목, 저자, 가격
# 형식으로 컬럼을 세팅해 csv로 뽑아보겠습니다.
# 가격에 "," 가 포함되어있으므로, 깨지지 않게 처리하기위해 ,를 없애고 txt파일로 써주셔야합니다.

f = codecs.open("c:/crawler2/result.txt", mode='w', encoding='utf-8')

# 데이터 추가
for i in range(0, len(title_list)):
    # 각 책에 대한 정보 write로 한 줄에 적기
    f.write(title_list[i].replace(",","") + "," +
            author_list[i].replace(",","") + "," +
            price_list[i].replace(",","") + "\n")   

# 파일 연결을 마지막에 끊어줘야 삭제나 수정, 이동가능
f.close()
```
