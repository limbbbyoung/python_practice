```python
# 크롤링 작업을 위한 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# YES24 접속

# browser 라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
browser= webdriver.Chrome('chromedriver.exe') # 확장자를 빼도 잘 돌아감

# .get(접속주소) 를 입력하면 해당 주소로 브라우저 창이 이동합니다.
browser.get("http://www.yes24.com/24/category/bestseller?categorynumber=001&sumgb=06&fetchsize=1000&pagenumber=1")

source = browser.page_source
print(type(source))
print(source)
```

```python
# 소스코드 파싱
yes24_code = BeautifulSoup(source, "html.parser")
print(type(yes24_code))
```

```python
# 순위별로 데이터를 가져오기 위해 처리
# 태그명이 div 이면서 클래스명이 ss_book_box인 요소를 가져와야함
book_detail_box = yes24_code.find_all("td", class_="goodsTxtInfo")
print(len(book_detail_box))

# 제목, 가격, 저자 정보를 저장할 리스트 생성
title_list = []
price_list = []
author_list = []
```

```python


    # browser 라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
    browser= webdriver.Chrome('chromedriver.exe') # 확장자를 빼도 잘 돌아감

    # .get(접속주소) 를 입력하면 해당 주소로 브라우저 창이 이동합니다.
    browser.get("http://www.yes24.com/24/category/bestseller?categorynumber=001&sumgb=06&fetchsize=1000&pagenumber=1")

    # 소스코드 저장
    source = browser.page_source
    
    # 파싱 
    yes24_code = BeautifulSoup(source, "html.parser")
    
    # 출력
    book_detail_box = yes24_code.find_all("td", class_="goodsTxtInfo")

    # 제목, 저자, 가격 가져오기
    for div in book_detail_box:
        p_list = div.find_all("p")
        book_author_list = div.find_all("div", class_="aupu")
        
        
        # 제목과 가격을 리스트에 추가
        print(p_list[0].text.strip())
        title_list.append(p_list[0].text.strip())
        print(p_list[1].text.strip())
        price_list.append(p_list[1].text.strip())
        
        # 저자를 리스트에 추가
        for item in book_author_list:
            print(item.text.strip().replace('\n',"").replace("\t","").replace(" ","").split("|")[0])
            author_list.append(item.text.strip().replace('\n',"").replace("\t","").split("|")[0])
            print("---------------------------------------------")
            
```

```python
# 위의 txt양식을 바꿔서 
# 제목, 저자, 가격
# 형식으로 컬럼을 세팅해 csv로 뽑아보겠습니다.

f = codecs.open("c:/crawler2/YES24best999.txt", mode='w', encoding='utf-8')

# 데이터 추가
for i in range(0, len(title_list)):
    # 각 책에 대한 정보 write로 한 줄에 적기
    f.write(title_list[i].replace(",","") + "," +
            author_list[i].replace(",","") + "," +
            price_list[i].replace(",","") + "\n")   

# 파일 연결을 마지막에 끊어줘야 삭제나 수정, 이동가능
f.close()
```
